```
# 로드한 파일의 데이터 타입 변경.
# 데이터의 과도한 메모리 사용으로 세션이 종료되는 것을 방지.

submission['ID'] = submission['ID'].astype(np.int16)
submission['item_cnt_month'] = submission['item_cnt_month'].astype(np.float32)

items['item_id'] = items['item_id'].astype(np.int16)
items['item_category_id'] = items['item_category_id'].astype(np.int16)

train['date_block_num'] = train['date_block_num'].astype(np.int16)
train['shop_id'] = train['shop_id'].astype(np.int16)
train['item_id'] = train['item_id'].astype(np.int16)
train['item_price'] = train['item_price'].astype(np.float32)
train['item_cnt_day'] = train['item_cnt_day'].astype(np.float32)
```

```
# 아웃 라이어 제거.
# 가격 및 판매 갯수가 너무 많은 인스턴스를 제거 또는 다른 값으로 대체.

median = train[(train.shop_id==32)&(train.item_id==2973)&(train.date_block_num==4)&(train.item_price>0)].item_price.median()
train.loc[train.item_price<0, 'item_price'] = median

train = train[train.item_price < 100000]
train = train[train.item_cnt_day < 1001]
```

```
#날짜 데이터 타입을 데이트타임으로.

import datetime

def to_datetime(column):
    d = column.split('.')
    return datetime.date(int(d[2]), int(d[1]), int(d[0]))

train['date'] = train['date'].apply(to_datetime)
```

```
#id 겹치는거 처리.

# Якутск Орджоникидзе, 56
train.loc[train.shop_id == 0, 'shop_id'] = 57
test.loc[test.shop_id == 0, 'shop_id'] = 57
# Якутск ТЦ "Центральный"
train.loc[train.shop_id == 1, 'shop_id'] = 58
test.loc[test.shop_id == 1, 'shop_id'] = 58
# Жуковский ул. Чкалова 39м²
train.loc[train.shop_id == 10, 'shop_id'] = 11
test.loc[test.shop_id == 10, 'shop_id'] = 11
```

```
from sklearn.preprocessing import LabelEncoder

#아이템, 샵 카테고리에서 서브타입 뽑아내는 코드.

shops.loc[shops.shop_name == 'Сергиев Посад ТЦ "7Я"', 'shop_name'] = 'СергиевПосад ТЦ "7Я"'
shops['city'] = shops['shop_name'].str.split(' ').map(lambda x: x[0])
shops.loc[shops.city == '!Якутск', 'city'] = 'Якутск'
shops['city_code'] = LabelEncoder().fit_transform(shops['city'])
shops = shops[['shop_id','city_code']]

cats['split'] = cats['item_category_name'].str.split('-')
cats['type'] = cats['split'].map(lambda x: x[0].strip())
cats['type_code'] = LabelEncoder().fit_transform(cats['type'])
# if subtype is nan then type
cats['subtype'] = cats['split'].map(lambda x: x[1].strip() if len(x) > 1 else x[0].strip())
cats['subtype_code'] = LabelEncoder().fit_transform(cats['subtype'])
cats = cats[['item_category_id','type_code', 'subtype_code']]

items.drop(['item_name'], axis=1, inplace=True)
```

```


from itertools import product

df_feat = []
cols = ['date_block_num','shop_id','item_id']
for i in range(34):
    sales = train[train.date_block_num==i]
    df_feat.append(np.array(list(product([i], sales.shop_id.unique(), sales.item_id.unique())), dtype='int16'))
    
df_feat = pd.DataFrame(np.vstack(df_feat), columns=cols)
df_feat['date_block_num'] = df_feat['date_block_num'].astype(np.int16)
df_feat['shop_id'] = df_feat['shop_id'].astype(np.int16)
df_feat['item_id'] = df_feat['item_id'].astype(np.int16)
df_feat.sort_values(cols,inplace=True)
```

```
# item, shop 카테고리 feature merge

items = pd.merge(items, cats, on = ['item_category_id'], how = 'left')
df_feat
```

```
# train의 샵, 아이템 아이디 별로 아이템 판매 갯수 group화.
# 일별 판매량으로 나와 있는 것을 달별 판매량으로 합치는 작업.

group = train.groupby(['date_block_num','shop_id','item_id']).agg({'item_cnt_day': ['sum']})
group.columns = ['item_cnt_month']
group.reset_index(inplace=True)

df_feat = pd.merge(df_feat, group, on=cols, how='left')
df_feat['item_cnt_month'] = (df_feat['item_cnt_month']
                                .fillna(0)
                                .clip(0,20) # NB clip target here
                                .astype(np.float16))
```

```
# test파일의 데이터타입 변경.

test['date_block_num'] = 34
test['date_block_num'] = test['date_block_num'].astype(np.int8)
test['shop_id'] = test['shop_id'].astype(np.int8)
test['item_id'] = test['item_id'].astype(np.int16)
test
```

```
# test를 df_feat에 concat후 nan값 처리.

df_feat = pd.concat([df_feat, test], ignore_index=True, sort=False, keys=cols)
df_feat.fillna(0, inplace=True) # 34 month
df_feat
```

```
# 가장 처음 팔린 달

train.sort_values('date_block_num', ascending = True, inplace = True)
df_first = pd.pivot_table(train, values = 'date_block_num', index = ['shop_id', 'item_id'], aggfunc = np.min)
df_first.reset_index(inplace = True)
df_first.columns = ['shop_id', 'item_id', 'shop_item_first_block']

df_temp = pd.pivot_table(train, values = 'date_block_num', index = 'item_id', aggfunc = np.min)
df_temp.reset_index(inplace = True)
df_temp.columns = ['item_id', 'item_first_block']
df_first = pd.merge(df_first, df_temp, on = 'item_id', how = 'left')

df_temp = pd.pivot_table(train, values = 'date_block_num', index = 'shop_id', aggfunc = np.min)
df_temp.reset_index(inplace = True)
df_temp.columns = ['shop_id', 'shop_first_block']
df_first = pd.merge(df_first, df_temp, on = 'shop_id', how = 'left')

del df_temp

df_feat = pd.merge(df_feat, df_first, on = ['shop_id', 'item_id'], how = 'left')
del df_first


# 가장 마지막에 팔린 달.
train.sort_values('date_block_num', ascending = True, inplace = True)
df_last = pd.pivot_table(train, values = 'date_block_num', index = ['shop_id', 'item_id'], aggfunc = np.max)
df_last.reset_index(inplace = True)
df_last.columns = ['shop_id', 'item_id', 'shop_item_last_block']

df_temp = pd.pivot_table(train, values = 'date_block_num', index = 'item_id', aggfunc = np.max)
df_temp.reset_index(inplace = True)
df_temp.columns = ['item_id', 'item_last_block']
df_last = pd.merge(df_last, df_temp, on = 'item_id', how = 'left')

df_temp = pd.pivot_table(train, values = 'date_block_num', index = 'shop_id', aggfunc = np.max)
df_temp.reset_index(inplace = True)
df_temp.columns = ['shop_id', 'shop_last_block']
df_last = pd.merge(df_last, df_temp, on = 'shop_id', how = 'left')

del df_temp

df_feat = pd.merge(df_feat, df_last, on = ['shop_id', 'item_id'], how = 'left')
del df_last

df_feat
```

```
# shop및 item_category에서 뽑아낸 type, city code를 df_feat에 merge

df_feat = pd.merge(df_feat, shops, on=['shop_id'], how='left')
df_feat = pd.merge(df_feat, items, on=['item_id'], how='left')
df_feat['city_code'] = df_feat['city_code'].astype(np.int8)
df_feat['item_category_id'] = df_feat['item_category_id'].astype(np.int8)
df_feat['type_code'] = df_feat['type_code'].astype(np.int8)
df_feat['subtype_code'] = df_feat['subtype_code'].astype(np.int8)

del shops
del items

df_feat
```

```
#뽑아낸 feature의 데이터 타입 변경.

df_feat['shop_item_first_block'] = df_feat['shop_item_first_block'].astype(np.float16)
df_feat['item_first_block'] = df_feat['item_first_block'].astype(np.float16)
df_feat['shop_first_block'] = df_feat['shop_item_first_block'].astype(np.float16)
df_feat['shop_item_last_block'] = df_feat['shop_item_last_block'].astype(np.float16)
df_feat['item_last_block'] = df_feat['item_last_block'].astype(np.float16)
df_feat['shop_last_block'] = df_feat['shop_last_block'].astype(np.float16)
```

```
# lag_feature [1, 2, 3, 5, 6, 11, 12]

# lag feature를 뽑아내는 코드.
# 이전 date_block의 값을 이번 date_block으로 가져와 merge하는 작업.

def lag_feature(df, lags, col):
  tmp = df[['date_block_num','shop_id','item_id',col]]
  for i in lags:
      shifted = tmp.copy()
      shifted.columns = ['date_block_num','shop_id','item_id', col+'_lag_'+str(i)]
      shifted['date_block_num'] += i
      df = pd.merge(df, shifted, on=['date_block_num','shop_id','item_id'], how='left')
  return df

# get_col은 내부적으로 lag_feature함수가 돌아감.
# index를 기준으로 val의 평균을 뽑아내며 col_name의 feature를 뽑아냄.
def get_col(df_feat, ind, val, lag_list, col_name):
    group = df_feat.groupby(ind).agg({val: ['mean']})
    group.columns = [col_name]
    group.reset_index(inplace=True)

    df_feat = pd.merge(df_feat, group, on=ind, how='left')
    df_feat[col_name] = df_feat[col_name].astype(np.float16)
    df_feat = lag_feature(df_feat, lag_list, col_name)
    df_feat.drop([col_name], axis=1, inplace=True)
    return df_feat

# get_col과 같으며 df_feat에는 item_price가 없으므로 train에서 뽑아옴.
def get_price_feat(df_feat, ind, val, lag_list, col_name):
    group = train.groupby(ind).agg({val: ['mean']})
    group.columns = [col_name]
    group.reset_index(inplace=True)

    df_feat = pd.merge(df_feat, group, on=ind, how='left')
    df_feat[col_name] = df_feat[col_name].astype(np.float16)
    df_feat = lag_feature(df_feat, lag_list, col_name)
    df_feat.drop([col_name], axis=1, inplace=True)
    return df_feat
```

```
# 정의한 함수의 사용.

lag_list = [1, 2, 3, 6]
df_feat = get_col(df_feat, ['date_block_num'], 'item_cnt_month', [1, 2, 6], 'date_item_mean_item_cnt')
df_feat = get_col(df_feat, ['date_block_num', 'shop_id', 'item_id'], 'item_cnt_month', lag_list, 'shop_item_mean_item_cnt')
df_feat = get_col(df_feat, ['date_block_num', 'shop_id'], 'item_cnt_month', [1, 2, 3], 'shop_mean_item_cnt')
df_feat = get_col(df_feat, ['date_block_num', 'item_id'], 'item_cnt_month', lag_list, 'item_mean_item_cnt')
df_feat = get_col(df_feat, ['date_block_num', 'item_category_id'], 'item_cnt_month', [1, 2, 3], 'cat_mean_item_cnt')
df_feat = get_col(df_feat, ['date_block_num', 'subtype_code'], 'item_cnt_month', [1, 2], 'sub_mean_item_cnt')
df_feat = get_col(df_feat, ['date_block_num', 'type_code'], 'item_cnt_month', [1, 2], 'type_mean_item_cnt')
```

```
# item_price를 이용한 feature 뽑아내기.
# index 기준으로 item_price의 median을 뽑아내고
# lag_feature에서 median 값을 빼 delta feature를 구함.

df_feat = get_price_feat(df_feat, ['date_block_num', 'shop_id', 'item_id'], 'item_price', [1, 6], 'shop_item_mean_item_price')
df_feat = get_price_feat(df_feat, ['date_block_num', 'item_id'], 'item_price', [1, 6], 'item_mean_item_price')
```

```
# delta 변수.
df_feat['delta_shop_item_price'] = df_feat['shop_item_mean_item_price_lag_1'] - df_feat['shop_item_mean_item_price_lag_6']
df_feat['delta_item_item_price'] = df_feat['item_mean_item_price_lag_1'] - df_feat['item_mean_item_price_lag_6']

df_feat['delta_item_item_cnt'] = df_feat['item_mean_item_cnt_lag_1'] - df_feat['item_mean_item_cnt_lag_3']
df_feat['delta_cat_item_cnt'] = df_feat['cat_mean_item_cnt_lag_1'] - df_feat['cat_mean_item_cnt_lag_3']
```

```
#month data

df_feat['month'] = df_feat['date_block_num'] % 12
days = pd.Series([31,28,31,30,31,30,31,31,30,31,30,31])
df_feat['days'] = df_feat['month'].map(days).astype(np.int8)

df_feat = df_feat[df_feat.date_block_num > 11]
```

```
# 데이터의 nan값 처리.

def fill_na(df):
    for col in df.columns:
        if ('_lag_' in col) & (df[col].isnull().any()):
            if ('item_cnt' in col):
                df[col].fillna(0, inplace=True)  

        if('first_block' in col):
          df[col].fillna(-1, inplace = True)

        if('last_block' in col):
          df[col].fillna(-1, inplace = True)
    return df

df_feat = fill_na(df_feat)
```
